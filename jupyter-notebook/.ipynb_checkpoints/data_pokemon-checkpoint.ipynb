### Considerando um exemplo correto de INSERT como:
    
### INSERT INTO pokemons (id, numero, nome, tipo1, tipo2, hp, ataque, defesa, especial, velocidade) VALUES ('sq_id_pokemons.nextval', 1, 'Bulbasaur', 'grass', 'poison', 253, 45, 49, 49, 65)

### Aqui, poderemos obter um facilitador para a criação de um script DML:

```python
for i in range(0, dados.shape[0]):
    print("INSERT INTO pokemons (id, numero, nome, tipo1, tipo2, hp, ataque, defesa, especial, velocidade) VALUES (sq_id_pokemons.nextval, " + str(dados.values[i][0]) + ", '" 
          + dados.values[i][1] + "', '" + dados.values[i][3] + "', '" 
          + dados.values[i][4] + "', " + str(dados.values[i][12]) + ", " 
          + str(dados.values[i][13]) + ", " + str(dados.values[i][14]) + ", " 
          +  str(dados.values[i][15]) + ", "+ str(dados.values[i][16]) + ");")
    

# Note que utilizei a mesma sintaxe da célula 3 deste arquivo para facilitar a compreensão.
```

### Para facilitar a importação das imagens em um projeto web, por exemplo, é possível utilizar o seguinte script:

```python
for i in range(0, dados.shape[0]):
    print("import pokemon" + str(dados.values[i][0])+ " from './assets/img/" + dados.values[i][1].lower() +".jpg'")
```
